```
import pandas as pd
import pyspc as cc
pesos=pd.read_csv("https://raw.githubusercontent.com/DanielValenciaL/Data/master/pesobalones.csv")
pesos=pesos.drop('Unnamed: 0',axis=1)
print(cc.spc(pesos)+cc.xbar_sbar()+cc.sbar())
print(cc.spc(pesos)+cc.xbar_rbar()+cc.rbar())
```
